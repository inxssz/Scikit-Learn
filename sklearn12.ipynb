Support Vector Machine Classification (SVM)

Konsep Dasar
Decision Boundary (Hyperplane)

- Ada 2 kelas, kelas hitam & putih

- Ada features X1 dan features X2

- Trus ada pemisahan, disebut jg sebagai decision boundary (Hyperplane)

- Garis yang paling baik memisahkan 2 kelas adalah garis yang memiliki margin paling besar.

- 1 Feature, decision boundarynya hanya berupa titik/nilai threshold

- 2 Feature (2 dimensi), decision boundarynya berupa garis lurus/linear

- 3 Feature, decision boundarynya berupa bidang datar atau disebut jg sebagai plane

- 4 atau lebih Feature, decision boundarynya berupa bidang multidimensi atau hyperplane

- Pada SVM setiap decision boundary yang dihasilkan disebut dengan hyperplane
